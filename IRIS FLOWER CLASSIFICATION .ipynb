```python
data_df=data.values
X=data_df[:,0:4]
Y=data_df[:,4]
print(X)
print(Y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
#print(Y_train)
```

# Model 1: SVM Algorithm

```python
from sklearn.svm import SVC
model_svc=SVC()
model_svc.fit(X_train,Y_train)
```

```python
prediction1=model_svc.predict(X_test)
from sklearn.metrics import accuracy_score
Accuracy=accuracy_score(Y_test,prediction1)
print("Accuracy: ",Accuracy*100)
for i in range(len(prediction1)):
    print(Y_test[i],prediction1[i])
```

# Model 2: Logistic Regression 

```python
from sklearn.linear_model import LogisticRegression 
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
model_lr=make_pipeline(StandardScaler(),LogisticRegression())
model_lr.fit(X_train,Y_train)
```

```python
prediction2=model_lr.predict(X_test)
from sklearn.metrics import accuracy_score
Accuracy=accuracy_score(Y_test,prediction2)
print("Accuracy: ",Accuracy*100)
for i in range(len(prediction2)):
    print(Y_test[i],prediction2[i])
```

# Model 3: Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
model_dt=DecisionTreeClassifier()
model_dt.fit(X_train,Y_train)
```

```python
prediction3=model_dt.predict(X_test)
from sklearn.metrics import accuracy_score
Accuracy=accuracy_score(Y_test,prediction3)
print("Accuracy: ",Accuracy*100)
for i in range(len(prediction3)):
    print(Y_test[i],prediction3[i])
```

```python
from sklearn.metrics import classification_report
print(classification_report(Y_test,prediction3))
```
